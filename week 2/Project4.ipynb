```python
# Program to determine Annual Tax Revenue (ATR)

# Input from user
experience = int(input("Enter years of work experience: "))
age = int(input("Enter age of staff: "))

# Determine ATR
if experience > 25 and age >= 55:
    atr = 5600000
elif experience > 20 and age >= 45:
    atr = 4480000
elif experience > 10 and age >= 35:
    atr = 1500000
elif experience < 10 and age < 35:
    atr = 550000
else:
    atr = 0  # If none of the conditions match

# Output result
print("Annual Tax Revenue (ATR): ₦", atr)
```
