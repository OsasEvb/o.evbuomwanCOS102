```python
import numpy as np

def solve_mission():
    print("--- Project II: Multi-Equation Solver ---")
    print("Which operation would you like to perform?")
    print("1. Solve Quadratic (Ax² + Bx + C = 0)")
    print("2. Solve Cubic (Ax³ + Bx² + Cx + D = 0)")
    print("3. Solve Quartic (Ax⁴ + Bx³ + Cx² + Dx + E = 0)")
    
    # Requesting the operation from the user
    choice = input("\nEnter choice (1, 2, or 3): ")

    # Handling inputs and logic based on the chosen operation
    if choice == '1':
        print("\nSolving Quadratic Equation...")
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        coeffs = [a, b, c]
        
    elif choice == '2':
        print("\nSolving Cubic Equation...")
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        coeffs = [a, b, c, d]
        
    elif choice == '3':
        print("\nSolving Quartic Equation...")
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        e = float(input("Enter E: "))
        coeffs = [a, b, c, d, e]
    else:
        print("Invalid operation selected.")
        return

    # Calculating roots
    roots = np.roots(coeffs)
    
    print("\nThe calculated roots are:")
    for i, r in enumerate(roots, 1):
        print(f"Root {i}: {r}")

if __name__ == "__main__":
    solve_mission()
```
