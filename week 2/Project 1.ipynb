## PSEUDOCODE
START
Input number_of_items
Set total = 0

FOR each item
    Input item_price
    total = total + item_price
END FOR

Display total

Input payment_amount

IF payment_amount >= total THEN
    change = payment_amount - total
    Display "Payment successful"
    Display "Change =", change
ELSE
    Display "Insufficient payment"
END IF

END
