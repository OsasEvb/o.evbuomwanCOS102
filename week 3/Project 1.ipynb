```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load dataset
data = pd.read_csv("GIG-logistics.csv")

def verify_employee():
    name = name_entry.get().strip()
    dept = dept_entry.get().strip()

    employee = data[(data['Name'].str.lower() == name.lower()) &
                    (data['Department'].str.lower() == dept.lower())]

    if len(employee) > 0:
        dept_members = data[data['Department'].str.lower() == dept.lower()]['Name'].tolist()

        if name in dept_members:
            dept_members.remove(name)

        members = "\n".join(dept_members)

        messagebox.showinfo(
            "Employee Verified",
            f"Welcome {name}!\n\nDepartment: {dept}\n\nOther members:\n{members}"
        )
    else:
        messagebox.showwarning(
            "Not Found",
            "Sorry, this employee does not exist."
        )

# GUI window
root = tk.Tk()
root.title("GIG Logistics Verification")
root.geometry("400x250")

tk.Label(root, text="Employee Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Department").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Verify Employee", command=verify_employee).pack(pady=10)

root.mainloop()
```
