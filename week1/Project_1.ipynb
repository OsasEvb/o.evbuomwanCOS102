```python
import math

# Example values
P = 1000      # Principal
R = 5.0       # Rate (%) per year
n = 12        # Compounds per year
T = 3         # Years
PMT = 100     # Annuity payment

# Simple Interest
A_simple = P * (1 + (R/100) * T)
print("Simple Interest:", A_simple)

# Compound Interest
A_comp = P * (1 + R/n) ** (n * T)
print("Compound Interest:", A_comp)

# Annuity Plan
rate_per_period = R / n
if rate_per_period == 0:
    A_annuity = PMT * n * T
else:
    A_annuity = PMT * (((1 + rate_per_period) ** (n * T) - 1) / rate_per_period)
print("Annuity Plan:", A_annuity)
```
